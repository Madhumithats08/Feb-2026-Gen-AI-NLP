## Task 1: Basic PromptTemplate

## Task 2: Multi-Input Prompt

```python
# Task 2

multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test cases
print(multi_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations

```python
# Task 3

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

```python
# Task 4

def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are a technical interviewer."
    elif role == "motivator":
        system_msg = "You are an inspiring mentor."
    else:
        system_msg = "You are an assistant."

    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Talk about {topic}")
    ])

    return chat_prompt.format_messages(topic=topic)

# Test
messages = get_chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(msg)
```

## Task 5: Input Validation

```python
# Task 5

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("kids", "funny"))
```

## Task 6: Prompt Generator Engine

```python
# Task 6

def generate_prompt(topic, audience, tone, style):

    # Validation
    audience, tone = validate_inputs(audience, tone)

    # Template selection
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} tone step by step"
    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"
    elif style == "storytelling":
        template = "Explain {topic} as a story for {audience} in a {tone} tone"
    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Reusability Test

```python
# Task 7

reuse_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

inputs = [
    ("AI", "beginners"),
    ("Python", "students"),
    ("Machine Learning", "engineers"),
    ("Data Science", "analysts"),
    ("Deep Learning", "researchers")
]

for topic, audience in inputs:
    print(reuse_template.format(topic=topic, audience=audience))
```

# Final Pipeline Flow

```python
def prompt_pipeline(topic, audience, tone, style):

    prompt = generate_prompt(topic, audience, tone, style)

    response = llm.invoke([("user", prompt)])

    return response.content


print(prompt_pipeline("Generative AI", "beginner", "fun", "teaching"))
```
